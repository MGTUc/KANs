```python
training_history1= pd.read_csv('./SS_KAN_KUL/test___model_saves_simple_2/training_history_Silverbox_state_[]_output_[]_batch_64_MLPKAN_subnet[12, 12]_1_100.csv')
training_history2= pd.read_csv('./SS_KAN_KUL/test___model_saves_simple_2/training_history_Silverbox_state_[]_output_[]_batch_64_EfficientKAN_grid5_1.csv')
training_history3= pd.read_csv('./SS_KAN_KUL/test___model_saves_simple_2/training_history_Silverbox_state_[]_output_[]_batch_64_FastKAN_grid8_1_100.csv')

raw_test_loss1 = pd.to_numeric(
    training_history1['test_loss'].astype(str).str.extract(r'([-+]?[0-9]*\.?[0-9]+(?:[eE][-+]?[0-9]+)?)')[0],
    errors='coerce'
)

raw_test_loss2 = pd.to_numeric(
    training_history2['test_loss'].astype(str).str.extract(r'([-+]?[0-9]*\.?[0-9]+(?:[eE][-+]?[0-9]+)?)')[0],
    errors='coerce'
)

raw_test_loss3 = pd.to_numeric(
    training_history3['test_loss'].astype(str).str.extract(r'([-+]?[0-9]*\.?[0-9]+(?:[eE][-+]?[0-9]+)?)')[0],
    errors='coerce'
)

valid_mask1 = raw_test_loss1.notna()
valid_mask2 = raw_test_loss2.notna()
valid_mask3 = raw_test_loss3.notna()
valid_epoch1 = training_history1.loc[valid_mask1, 'time']
valid_test_loss1 = raw_test_loss1.loc[valid_mask1]

valid_epoch2 = training_history2.loc[valid_mask2, 'time']
valid_test_loss2 = raw_test_loss2.loc[valid_mask2]

valid_epoch3 = training_history3.loc[valid_mask3, 'time']
valid_test_loss3 = raw_test_loss3.loc[valid_mask3]

palette = {
    'RS-KAN': '#1F6EB3',
    'FastKAN': '#DE2B25',
    'EfficientKAN': '#FC9070',
}

# plt.plot(training_history1['time'], training_history1['train_loss'], label='train_loss')
plt.plot(valid_epoch1, valid_test_loss1, linestyle='-', label='RS-KAN', color=palette['RS-KAN'])
# plt.plot(training_history2['time'], training_history2['train_loss'], label='train_loss2')
plt.plot(valid_epoch2, valid_test_loss2, linestyle='-', label='EfficientKAN', color=palette['EfficientKAN'])
# plt.plot(training_history3['time'], training_history3['train_loss'], label='train_loss3')
plt.plot(valid_epoch3, valid_test_loss3, linestyle='-', label='FastKAN', color=palette['FastKAN'])
plt.xlabel('Time (s)')
plt.ylabel('MSE')
# plt.yscale('log')
plt.ylim(0, 1e-3)
plt.title('MSE loss over time on the normalized Silverbox dataset')
plt.legend()
plt.grid()
plt.savefig("./SS_KAN_KUL/test___model_saves_simple_2/acc2.pdf", bbox_inches="tight", facecolor="white")
plt.show()
```
