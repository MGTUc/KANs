```python
training_history20_SilU = pd.read_csv('./SS_KAN_KUL/test___model_saves_simple_2/training_history_Silverbox_state_[]_output_[]_batch_128_MLPKAN_subnet[32, 32]_1.csv')

raw_test_loss = pd.to_numeric(
    training_history20_SilU['test_loss'].astype(str).str.extract(r'([-+]?[0-9]*\.?[0-9]+(?:[eE][-+]?[0-9]+)?)')[0],
    errors='coerce'
)

valid_mask = raw_test_loss.notna()
valid_time = training_history20_SilU.loc[valid_mask, 'time']
valid_test_loss = raw_test_loss.loc[valid_mask]

plt.plot(training_history20_SilU['time'], training_history20_SilU['train_loss'], label='train_loss')
plt.plot(valid_time, valid_test_loss, linestyle='-', label='test_loss')
plt.xlabel('Time (s)')
plt.ylabel('loss')
plt.yscale('log')
plt.title('Model Performance Over Time')
plt.legend()
plt.show()
```
