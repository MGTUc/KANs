```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load CSVs (same sources as the notebook)
MLPKAN_restrainable = pd.read_csv('./parameterTests/MLPKAN_restrainable.csv')
MLPKAN_res = pd.read_csv('./parameterTests/MLPKAN_res.csv')
MLPKAN_nores = pd.read_csv('./parameterTests/MLPKAN_nores.csv')
FastKAN = pd.read_csv('./parameterTests/FastKAN2.csv')
EfficientKAN = pd.read_csv('./parameterTests/EfficientKAN2.csv')

methods = {
    'RS-KAN (trainable residual)': (MLPKAN_restrainable, {'Layers': 2, 'Nodes': 12}),
    'RS-KAN (static residual)': (MLPKAN_res, {'Layers': 2, 'Nodes': 12}),
    'RS-KAN (no residual)': (MLPKAN_nores, {'Layers': 2, 'Nodes': 12}),
    'FastKAN': (FastKAN, {'Grid Size': 8}),
    'EfficientKAN': (EfficientKAN, {'Grid Size': 5}),
}

levels = ['Easy', 'Medium', 'Hard']
rows = []

for level in levels:
    # compute common functions per level across all methods (with R2>=0.999 and method-specific filters)
    common_sets = []
    for name, (df, extra_filters) in methods.items():
        cond = (df['R2 Score (Test)'] >= 0.999) & (df['Level'] == level)
        for k, v in extra_filters.items():
            cond &= (df[k] == v)
        funcs = set(df[cond]['Function'].tolist())
        common_sets.append(funcs)
    common = set.intersection(*common_sets) if common_sets else set()

    # gather times for each method restricted to common functions
    for name, (df, extra_filters) in methods.items():
        cond = (df['R2 Score (Test)'] >= 0.999) & (df['Level'] == level) & (df['Function'].isin(common))
        for k, v in extra_filters.items():
            cond &= (df[k] == v)
        times = df[cond]['time'].tolist()
        for t in times:
            rows.append({'Method': name, 'Level': level, 'time': t})

# Build DataFrame for plotting
plot_df = pd.DataFrame(rows)
if plot_df.empty:
    raise SystemExit('No common functions found across methods for any level. Nothing to plot.')

# Custom palette: different shades of green for MLPKAN, blue for FastKAN, orange for EfficientKAN
palette = {
    'RS-KAN (trainable residual)': '#1F6EB3',
    'RS-KAN (static residual)':           '#57A0CE',
    'RS-KAN (no residual)':        '#A3CCE3',
    
    'FastKAN':                '#DE2B25',
    'EfficientKAN':           '#FC9070',
}

plt.figure(figsize=(18, 6))
# Show grouped boxes per Level with Method as hue
sns.boxplot(x='Level', y='time', hue='Method', data=plot_df, palette=palette, showmeans=True, meanprops=dict(marker='+', markerfacecolor='black', markeredgecolor='black', markersize=5))
plt.ylabel('Time (seconds)')
plt.ylim(0,10)
plt.title('Time to solve common equations (R² > 0.999) — all levels and methods', fontweight='medium', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Method')
plt.tight_layout()
plt.savefig("./parameterTests/speed_comparison.pdf", bbox_inches="tight", facecolor="white")
plt.show()
```
