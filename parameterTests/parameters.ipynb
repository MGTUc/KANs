```python
from efficient_kan import KAN as efficientKAN
from mlpkan.MLPKAN import MLPKAN
from fastkan import FastKAN
from mlp.MLP import standardMLP
```

```python
fastkan = FastKAN([2, 3, 1], num_grids=8)
efficientkan = efficientKAN([2, 3, 1], grid_size=5, spline_order=3)
mlpkan = MLPKAN([2, 3, 1], subnetwork_shape=[5,3])
```

```python
def count_params(model):
    total = sum(p.numel() for p in model.parameters())
    trainable = sum(p.numel() for p in model.parameters() if p.requires_grad)
    return total, trainable

# Match your runner-style architecture
in_dim = 2
layers = [in_dim, 3, 1]

models = {
    "efficientKAN": efficientKAN(layers),
    "FastKAN": FastKAN(layers),
    "MLPKAN": MLPKAN(layers, subnetwork_shape=[5]),
    "MLP": standardMLP([2,8,8,1]),
}

for name, model in models.items():
    total, trainable = count_params(model)
    print(f"{name:12s} total={total:8d}  trainable={trainable:8d}")
```
